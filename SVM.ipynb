```
import numpy as np
import time
from sklearn.svm import LinearSVC
from sklearn.svm import SVC

def loaddata(pathin):
    def convert_dense_vector(text,dim):
        v=np.zeros(dim)
        x_fea=text.split()
        for x in x_fea:
            fea=x.split(':')
            v[int(fea[0])]=float(fea[1])
        return v

    with open(pathin,'r') as f:
        d_lines=f.read().splitlines()
  
    with open("/content/drive/My Drive/Project2/Datasets/words_idf",'r') as f:
        dim=f.read().splitlines()

    X=[]
    Y=[]
    for doc in d_lines:
        fea_doc=doc.split('<<>>')
        Y.append(int(fea_doc[0]))
        X.append(convert_dense_vector(fea_doc[2],dim))
    
    return X,Y

# accuracy function
def accuracy(pre,Y_test):
    pre_true=np.equal(pre,Y_test)
    return float(pre_true.sum()/len(pre))

#load data
X_train,Y_train=loaddata("/content/drive/My Drive/Project2/Datasets/train_tfidf_df=3")
X_test,Y_test=loaddata("/content/drive/My Drive/Project2/Datasets/test_tfidf")

```

```
#linear SVM
t=time.time()
model=LinearSVC(tol=1e-4,C=10)
model.fit(X_train,Y_train)
print("time train=",time.time()-t,end=';')
pre_test=model.predict(X_test)
pre_train=model.predict(X_train)
print('pre test=',str(accuracy(pre_test,Y_test)),'; pre train=',str(accuracy(pre_train,Y_train)))
```

```
# kernel SVM
t=time.time()
model=SVC(tol=1e-3,C=50.0,gamma=0.1,kernel='rbf')
model.fit(X_train,Y_train)
print("time train=",time.time()-t,end=';')
pre_test=model.predict(X_test)
pre_train=model.predict(X_train)
print('pre test=',str(accuracy(pre_test,Y_test)),'; pre train=',str(accuracy(pre_train,Y_train)))
```
